## Шаг 3. Поиск корреляции между числовыми признаками

Вычислите корреляции по Пирсону и по Спирмену между всеми парами числовых признаков. Подумайте, в чем между ними разница?

Подсказка: вы можете вычислять корреляции вручную, это будет даже похвально. А можете использовать встроенный метод `df_numeric.corr(...)`. Рекомендуется посмотреть документацию перед использованием: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html

```python
### BEGIN YOUR CODE
import numpy as np

def calculate_pearson_correlation(df):
    corr_matrix = pd.DataFrame(index=df.columns, columns=df.columns)

    for col1 in df.columns:
        for col2 in df.columns:
            x = df[col1].dropna()
            y = df[col2].dropna()

            common_idx = x.index.intersection(y.index)
            if len(common_idx) < 2:
                corr_matrix.loc[col1, col2] = np.nan
                continue

            x_vals = x.loc[common_idx]
            y_vals = y.loc[common_idx]

            numerator = ((x_vals - x_vals.mean()) * (y_vals - y_vals.mean())).sum()
            denominator = np.sqrt(((x_vals - x_vals.mean())**2).sum() *
                                  ((y_vals - y_vals.mean())**2).sum())

            if denominator == 0:
                corr_matrix.loc[col1, col2] = np.nan
            else:
                corr_matrix.loc[col1, col2] = numerator / denominator

    return corr_matrix.astype(float)

def calculate_spearman_correlation(df):
    corr_matrix = pd.DataFrame(index=df.columns, columns=df.columns)

    for col1 in df.columns:
        for col2 in df.columns:
            x = df[col1].dropna()
            y = df[col2].dropna()

            common_idx = x.index.intersection(y.index)
            if len(common_idx) < 2:
                corr_matrix.loc[col1, col2] = np.nan
                continue

            x_vals = x.loc[common_idx]
            y_vals = y.loc[common_idx]

            x_ranks = x_vals.rank(method='average')
            y_ranks = y_vals.rank(method='average')

            # Вычисляем корреляцию Пирсона между рангами
            x_mean = x_ranks.mean()
            y_mean = y_ranks.mean()

            numerator = ((x_ranks - x_mean) * (y_ranks - y_mean)).sum()
            denominator = np.sqrt(((x_ranks - x_mean)**2).sum() *
                                  ((y_ranks - y_mean)**2).sum())

            if denominator == 0:
                corr_matrix.loc[col1, col2] = np.nan
            else:
                corr_matrix.loc[col1, col2] = numerator / denominator

    return corr_matrix.astype(float)

pearson_corrs = calculate_pearson_correlation(df_numeric)
spearman_corrs = calculate_spearman_correlation(df_numeric)

print("Корреляция Пирсона:")
print(pearson_corrs)
print("\nКорреляция Спирмена:")
print(spearman_corrs)

### END YOUR CODE
```

## Шаг 4. Вычисление статистик

Для каждого числового признака вычислите среднее, медиану и стандартное отклонение.

Результаты запишите в соответствующие словари `means`, `medians` и `stds`, в которых ключ (`key`) - название признака (как он записан в датафрейме), значение (`value`) - статистика.

```python
### BEGIN YOUR CODE
means = {}
medians = {}
stds = {}

for column in df_numeric.columns:
    means[column] = df_numeric[column].mean()
    medians[column] = df_numeric[column].median()
    stds[column] = df_numeric[column].std()

print("Средние значения (means):")
for col, val in means.items():
    print(f"  {col}: {val:.4f}")

print("\nМедианы (medians):")
for col, val in medians.items():
    print(f"  {col}: {val:.4f}")

print("\nСтандартные отклонения (stds):")
for col, val in stds.items():
    print(f"  {col}: {val:.4f}")
### END YOUR CODE
```

## Шаг 5. Стандартизация признаков

Выполните стандартизацию для каждого числового признака:

$$
z = \dfrac{x - \mu}{\sigma}
$$

где $\mu$ - среднее по признаку, $\sigma$ - стандартное отклонение

```python
### BEGIN YOUR CODE
df_standart = df_numeric.copy()

for column in df_standart.columns:
    mu = means[column]
    sigma = stds[column]

    if sigma != 0:
        df_standart[column] = (df_standart[column] - mu) / sigma
    else:
        df_standart[column] = 0
print(df_standart.head())
### END YOUR CODE
```

# Задание 4. Поиск выбросов

В этом задании будем использовать полученные ранее датафреймы `df_standart` и `df_ohe`.

Очистите данные от выбросов, используя метод $3 \sigma$ и `df_standart`. Поскольку числовые данные уже стандартизированы, выбросом будем считать объект $|x| > 3$ (длина объекта больше 3).

Вам необходимо найти выбросы среди `df_standart`, удалить их, если они есть, и удалить соответствующие объекты среди `df_ohe`. Объедините датафреймы и запишите результат в переменную `df_filtered`.

**ВАЖНО**: не перезаписывайте переменные `df_standart` и `df_ohe`. Для промежуточных результатов создавайте другие переменные, например `df_standart_f` и `df_ohe_f`.

```python
### BEGIN YOUR CODE
outlier_mask = (df_standart.abs() > 3).any(axis=1)

print("Статистика выбросов:")
print(f"Всего объектов: {len(df_standart)}")
print(f"Выбросов: {outlier_mask.sum()}")
print(f"Процент выбросов: {outlier_mask.sum() / len(df_standart) * 100:.2f}%")

df_standart_f = df_standart[~outlier_mask].copy()  # f от "filtered"
df_ohe_f = df_ohe[~outlier_mask].copy()

print(f"\nПосле удаления выбросов:")
print(f"  df_standard_f размер: {df_standart_f.shape}")
print(f"  df_ohe_f размер: {df_ohe_f.shape}")

df_filtered = pd.concat([df_standart_f, df_ohe_f], axis=1)

print(f"\nОбъединенный df_filtered размер: {df_filtered.shape}")
### END YOUR CODE
```
